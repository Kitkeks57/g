```python
def iter(A, B, c, sch):

    minim = sred_otk(A, 0, sch, c)

    k = 0

    for i in range(len(A[0, :])):

        if sred_otk(A, i, sch, c) > minim:

            minim = sred_otk(A, i, sch, c)

            k = i

    i = 0

    j = 0

    while i < len(A[0, :]) and j == 0:

        if B[k, i] == 1 and suma(A, i, k, sch, c) < sred_otk(A, k, sch, c):

            A[:, k] = A[:, k] + A[:, i]

            A = np.delete(A, i, axis=1)

            B[:, k] = (B[:, k] + B[:, i])%2

            B[k, :] = (B[k, :] + B[i, :])%2

            B = np.delete(B, i, axis=0)

            B = np.delete(B, i, axis=1)

            c[k] = c[k] + ' ' + c[i]

            c.remove(c[i])

            j = j + 1

        i = i + 1

    if j == 0:

        A = np.delete(A, k, axis=1)

        B = np.delete(B, k, axis=0)

        B = np.delete(B, k, axis=1)

        print(c[k])

        c.remove(c[k])

        sch += 1

    return [A, B, c, sch]
```

```python
def sred_otk(A, n, sch, c):

    sum = 0

    for i in range(6):

        sum += (A[i, n] - 100/(len(c) + sch))**2

    return sum
```

```python
def suma(A, i, k, sch, c):

    l = A[:, i] + A[:, k]

    sum = 0

    for i in range(6):

        sum += (l[i] - 100/(len(c) + sch))**2

    return sum
```
